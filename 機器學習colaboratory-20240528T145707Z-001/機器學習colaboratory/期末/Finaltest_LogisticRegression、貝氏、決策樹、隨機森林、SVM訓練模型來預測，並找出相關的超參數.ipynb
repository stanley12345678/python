# LogisticRegression

```python
"""資料都填上平均值"""
age_mean = data['age'].mean()
data['age'] = data['age'].fillna(age_mean)
fare_mean = data['fare'].mean()
data['fare'] = data['fare'].fillna(fare_mean)
"""登船港口embarked 轉換"""
x_OneHot_data = pd.get_dummies(data=data,columns=["embarked" ])
```

```python
feature_name = ['age', 'fare', 'survived']
```

```python
X = data[feature_name]
X
```

```python
y = data ['pclass']
y
```

```python
from sklearn.model_selection import train_test_split
X_train_raw,X_test_raw,y_train,y_test=train_test_split(data["survived"].astype("U"),data["age"].astype("U")) #超參數
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect=TfidfVectorizer()
X_train=vect.fit_transform(X_train_raw)
print(X_train[10])
```

```python
from sklearn.linear_model import LogisticRegression
cls=LogisticRegression()
cls.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
score1=cross_val_score(cls,X_train,y_train,cv=5,scoring="accuracy")
```

```python
print("accuracies:%s"%score1)
print("mean accuracy:%s"%np.mean(score1))
```

```python
import pandas as pd 
```

```python
data = pd.read_csv("/content/drive/MyDrive/機器學習colaboratory/期中/Marketing_Data.csv")
data
```

```python
feature_name = ['youtube', 'facebook', 'newspaper']  #新增陣列
```

```python
import seaborn as sns
```

```python
sns.pairplot(data, x_vars=feature_name, y_vars='sales', kind='reg')
```

```python
X = data[feature_name]
X
```

```python
y = data ['sales']
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

model.score(X_test, y_test)
```

```python
model.predict([[100,50,20]])
```

```python
import pandas as pd  #載入pandas套件
import numpy as np  #載入numpy套件
```

```python
sms = pd.read_csv("/content/drive/MyDrive/機器學習colaboratory/ch7垃圾分類/sms.csv") #用pandas讀csv
sms.head() #列出前五列
```

```python
from sklearn.model_selection import train_test_split  #載入sklearn的切割套件
X_train_raw, X_test_raw, y_train, y_test = train_test_split(sms["message"], sms["label"]) #75%訓練25%測試
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vect = TfidfVectorizer()
X_train = vect.fit_transform(X_train_raw)
```

```python
print(X_train[3000])
```

```python
print(X_train[3000])
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
cls = LogisticRegression()
cls.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score1 = cross_val_score(cls, X_train, y_train, cv=5, scoring="accuracy")
```

```python
print("accuracies: %s" % score1)
```

```python
print("mean accuracies: %s" % np.mean(score1))
```

```python
X_test = vect.transform(X_test_raw)
score2 = cross_val_score(cls,X_test,y_test,cv=5,scoring="accuracy")
```

```python
print("mean accuracies: %s"% np.mean(score2))
```

```python
score3 = cross_val_score(cls,X_test,y_test,cv=5,scoring="precision")
print("mean accuracies: %s"% np.mean(score3))
```

```python
score4 = cross_val_score(cls,X_test,y_test,cv=5,scoring="recall")
print("mean recall: %s"% np.mean(score4))
```

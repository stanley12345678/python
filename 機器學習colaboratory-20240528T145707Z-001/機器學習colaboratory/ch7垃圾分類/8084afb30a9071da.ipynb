```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/20220221Learning/dataset/sms.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train_raw , X_test_raw , y_train , y_test = train_test_split(df["message"],df["label"])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vect = TfidfVectorizer()
X_train =vect.fit_transform(X_train_raw)
```

```python
print(X_train[3000])
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
cls = LogisticRegression()
cls.fit( X_train , y_train )
```

```python
from sklearn.model_selection import cross_val_score
```

```python
X_test = vect.transform(X_test_raw)
```

X_train , y_train score

```python
scorel = cross_val_score(cls , X_train , y_train , cv = 5 , scoring = "accuracy")
```

```python
print("accuracies: %s" % scorel)
```

```python
print ("accuracies: %s" % np.mean(scorel))
```

```python
X_test = vect.transform(X_test_raw)
```

X_test , y_test score

```python
scorel = cross_val_score(cls , X_test , y_test , cv = 5 , scoring = "accuracy")
```

```python
print("accuracies: %s" % scorel)
```

```python
print ("accuracies: %s" % np.mean(scorel))
```

```python
score2 = cross_val_score(cls , X_test , y_test , cv = 5 , scoring = "accuracy")
print ("accuracies: %s" % np.mean(score2))
```

```python
score3 = cross_val_score(cls , X_test , y_test , cv = 5 , scoring = "precision")
print ("precision: %s" % np.mean(score3))
```

```python
score3 = cross_val_score(cls , X_test , y_test , cv = 5 , scoring = "recall")
print ("recall: %s" % np.mean(score3))
```

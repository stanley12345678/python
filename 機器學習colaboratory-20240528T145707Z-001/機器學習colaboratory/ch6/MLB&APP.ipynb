```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/機器學習colaboratory/期中/Marketing_Data.csv")
```

```python
data = np.array(df.drop(labels="sales",axis=1))
result = np.array(df["sales"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(data,result,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.predict([[100,60,30]])
```

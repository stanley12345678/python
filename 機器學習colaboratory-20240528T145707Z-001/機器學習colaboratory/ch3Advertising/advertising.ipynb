```python
import pandas as pd  #載入pandas函式庫
```

```python
data = pd.read_csv("/content/drive/MyDrive/機器學習colaboratory/Advertising/Advertising.csv 的副本", index_col = 0)  #用pandas讀CSV資料集
data
```

```python
feature_name = ['TV', 'Radio', 'Newspaper']  #新增陣列
```

```python
import seaborn as sns  #載入seaborn函式庫
```

```python
sns.pairplot(data, x_vars=feature_name, y_vars='Sales', kind='reg')  #使用seaborn繪圖
```

```python
X = data[feature_name]
X
```

```python
y = data ['Sales']
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

model.score(X_test, y_test)
```

```python
model.predict([[100,50,20]])
```
